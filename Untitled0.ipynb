```python
import pandas as pd
sales1 = pd.DataFrame({'Дата':['2022-01-01','2022-01-02','2022-01-03'], 'Выручка':[1000,2000,1500]})
sales2 = pd.DataFrame({'Дата':['2022-01-01','2022-01-02','2022-01-03'], 'Выручка':[1500,2500,1800]})
total_sales = pd.merge(sales1,sales2,on= 'Дата', suffixes=('_магазин1', '_магазин2'))
total_sales['Общая выручка']= total_sales['Выручка_магазин1'] + total_sales['Выручка_магазин2']

result = total_sales[['Дата','Общая выручка']]
print(result)
```

```python
import pandas as pd
sales = pd.DataFrame({'product': ['product_1','product_2','product_3','product_4'],
'month_1':[100,200,300,400],
'month_2':[200,250,150,300],
'month_3':[150,300,100,200],
})

conversion_rate = pd.Series({'product_1': 0.1,'product_2': 0.75,'product_3': 0.55,'product_4': 0.30,})
total_conversion = sales.set_index('product').multiply(conversion_rate,axis=0).sum(axis=1)

sales['total_conversion'] = total_conversion.values
print(sales)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 6))

# Первый график
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title('График sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Второй график
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title('График cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')

# Третий график
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title('График tan(x)')
plt.xlabel('x')
plt.ylabel('tan(x)')

# Четвертый график
plt.subplot(2, 2, 4)
plt.plot(x, x)
plt.title('График y=x')
plt.xlabel('x')
plt.ylabel('y')

plt.suptitle('Несколько графиков на одной фигуре')
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
x, y = np.meshgrid(x, y)
z = np.sin(x) - y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)

ax.set_title('3D график f(x, y) = sin(x) - y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Значения x
x = np.linspace(-10, 10, 200)

# Уравнения
y1 = x + 1       # a = 1, b = 1
y2 = 2 * x + 1   # a = 2, b = 1
y3 = x + 2       # a = 1, b = 2

# Построение графиков
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='a=1, b=1')
plt.plot(x, y2, label='a=2, b=1')
plt.plot(x, y3, label='a=1, b=2')

plt.title('Графики для различных уравнений')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5, ls='--')  # Горизонтальная линия y=0
plt.axvline(0, color='black', lw=0.5, ls='--')  # Вертикальная линия x=0
plt.legend()
plt.grid()
plt.show()
```
